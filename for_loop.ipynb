Given a list of words ['Python', 'Java', 'C++', 'JavaScript', 'Ruby'], find the longest word using a for loop.


```python
words=['Python', 'Java', 'C++', 'JavaScript', 'Ruby']
len_word=words[0] # Initialize len_word with the first word
for i in words:
   if len(i) > len(len_word): # Compare the length of current word with the length of the longest word found so far
     len_word=i
print(len_word) # Print the longest word after the loop finishes
```

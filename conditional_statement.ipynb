Question 13: Password Strength Checker
Write a Python program that takes a password as input and checks its strength:
•	If the length is less than 6 characters, print "Weak password."
•	If the length is 6 or more, check:
o	If it contains numbers and special characters, print "Strong password."
o	Else, print "Moderate password, add special characters."


```python
password=input("enter the password : ")
if len(password)<6:
  print("Weak Password")
elif len(password)>=6:
  if password.isdigit() and password.punctuation():
    print("Strong Password")
  else:
    print("Moderate password, add special characters.")
```
